###  1.Data Cleaning 

Cleaning dataset to treat missing values and correcting data types 

```python
df.head()
```

## Taking only necessary column for analysis

```python
df = df[['Make','Model','Ex-Showroom_Price','Displacement', 'Valves_Per_Cylinder', 
             'Engine_Type', 'Fuel_Tank_Capacity',  'City_Mileage',  'Seating_Capacity',
              'Child_Safety_Locks',  'Parking_Assistance',  'Navigation_System']]
df.head()
```

 1.Ex-Showroom Price:  Removing the unnecessary strings from Ex-Showroom_Price column 

```python
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace('Rs. ','')
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace(',','')
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace(' ','')
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].astype('int64')
```

```python
df.rename(columns= {'Make': 'Company'}, inplace= True)
df.rename(columns= {'Ex-Showroom_Price': 'price'}, inplace= True)
df.rename(columns= {'Valves_Per_Cylinder': 'Cylinders'}, inplace= True)
df.head()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['Company'].unique()
```

2. we have to drop all NaN values from comapany, displacement columns as it is categorical and cannot be replaced 

```python
df['Company'] = df['Company'].replace(np.nan,'Unspecified')
df['Displacement'] = df['Displacement'].replace(np.nan,'Unspecified')

df['Displacement'].isna().sum()
```

3. Cleaning the fuel tank capacity column

```python
df['Fuel_Tank_Capacity'].unique()
```

```python
#  to remove the NaN values we can replace these with median values. but first have to convert it in numeric data type
df['Fuel_Tank_Capacity']= df['Fuel_Tank_Capacity'].str.replace(' litres','')
df['Fuel_Tank_Capacity'].astype('float')
df['Fuel_Tank_Capacity'].median()
df['Fuel_Tank_Capacity'] = df['Fuel_Tank_Capacity'].fillna(df['Fuel_Tank_Capacity'].median())
```

```python
df['Fuel_Tank_Capacity'].isna().sum()
```

4. cleaning the City_Mileage column

```python
df['City_Mileage'].head()
```

```python
df['City_Mileage'].isna().sum()
```

```python
#  to remove the NaN values we can replace these with median values. but first have to convert it in numeric data type
df['City_Mileage']= df['City_Mileage'].str.replace(' km/litre','')
df['City_Mileage']= df['City_Mileage'].str.replace('?','')
df['City_Mileage']= df['City_Mileage'].str.replace(',','.') # as some values have , at the place of .
df['City_Mileage']= df['City_Mileage'].str.replace('12.5-12.7','12.5') # as some values have , at the place of .
df['City_Mileage']= df['City_Mileage'].str.replace('26032','26.3')

df['City_Mileage'].unique()
df['City_Mileage']= df['City_Mileage'].astype(float)
df['City_Mileage'] = df['City_Mileage'].fillna(df['City_Mileage'].median())
df['City_Mileage'].unique()
```

5. Cleaning Seating_Capacity column

```python
df['Seating_Capacity'].isna().sum()
```

```python
# just dropping 6 null values
df['Seating_Capacity'] = df['Seating_Capacity'].replace(np.nan,'0')

df['Seating_Capacity'].isna().sum()
df['Seating_Capacity'] = df['Seating_Capacity'].astype('int64')
```

6. Cleaning Child safety locks column

```python
df['Child_Safety_Locks'].isna().sum()
```

```python
df['Child_Safety_Locks'] = df['Child_Safety_Locks'].replace(np.nan,'Unspecified')
df['Child_Safety_Locks'].isna().sum()
```

7. Cleaning parking Assistance column

```python
df['Parking_Assistance'] = df['Parking_Assistance'].replace(np.nan,'No')
df['Parking_Assistance'].isna().sum()
```

8. Cleaning Navigation_System column


```python
df['Navigation_System'] = df['Navigation_System'].replace(np.nan,'No') # replacing NaN with no
df['Navigation_System'].unique()
```

9. Cleaning Engine type column

```python
df['Engine_Type'] = df['Engine_Type'].replace(np.nan,'Unspecified')
df['Engine_Type'].isna().sum()
```

10. Cleaning Cylinders column

```python
df['Cylinders'] = df['Cylinders'].fillna(df['Cylinders'].median())
df['Cylinders'].isna().sum()
```

```python
df['Displacement'] = df['Displacement'].str.replace("cc","")
df['Displacement']= df['Displacement'].str.replace('Unspecified','0')
df['Displacement'] = df['Displacement'].astype('int64')
```
