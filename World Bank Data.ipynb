```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as py
```

```python
population = pd.read_csv("E:\Rayyan-DS\JupiterNotebook\world population data project\country_population.csv")

population
```

```python
population.drop(columns = ['Indicator Name','Indicator Code'], axis=1, inplace = True)
```

```python
population
```

```python
population.isnull().sum()
```

```python
#  as we have missing values for various years and we cant handle population by any statistical means so we are dropping it.
population.dropna()
```

```python
# creating years array fro later to merge in dataframe
years = [str(i) for i in range(1960,2017)]
years
```

## Melting the dataframe for better representation



```python
#### melting a dataframe is to unpivot the dataframe 
##### Reversing the columns and rows
population = population.melt(id_vars = "Country Name", value_vars = years, var_name ='year', value_name = 'populations')
```

```python
population.head()
```

```python
population.sort_values(by = ['Country Name','year'])
```

##  loading life expectancy dataset

```python
life_Exp = pd.read_csv("E:\Rayyan-DS\JupiterNotebook\world population data project\life_expectancy.csv")
life_Exp 
life_Exp.drop(columns = ['Indicator Name','Indicator Code'])
life_Exp.dropna()
life_Exp = life_Exp.melt(id_vars = "Country Name", value_vars = years, var_name ='year', value_name = 'Life Expectancy')
```

```python
life_Exp.dropna()
```

```python
life_Exp.sort_values(by = ['Country Name','year'])
```

## Loading fertility dataset

```python
fert_rate = pd.read_csv(r"E:\Rayyan-DS\JupiterNotebook\world population data project\fertility_rate.csv")
fert_rate.head()
```

```python
fert_rate.drop(columns = ['Indicator Name','Indicator Code'])
fert_rate = fert_rate.melt(id_vars = "Country Name", value_vars = years, var_name ='year', value_name = 'Fertility Rate')
```

```python
fert_rate.dropna()
```

```python
fert_rate.sort_values(by = ['Country Name','year'])
```

## Merging all three dataframes


```python
# Merging population and life exp dataframes

population = population.merge(life_Exp, how ='inner', on =['Country Name', 'year'])
#  merging fertility rate dataframe

population = population.merge(fert_rate, how ='inner', on =['Country Name', 'year'])
```

```python
meta_data = pd.read_csv("E:\Rayyan-DS\JupiterNotebook\world population data project\Metadata_Country.csv")
meta_data
```

```python
meta_data  = meta_data.drop("Unnamed: 5", axis=1) ## dropping unnamed column
meta_data.head()
```

```python
meta_data = meta_data.drop(columns = ['SpecialNotes','IncomeGroup'], axis = 1 )
```

```python
meta_data.rename(columns = {'TableName':'Country Name'}, inplace = True) 
meta_data.rename(columns = {'Country Name\t':'Country Name'}, inplace = True) 
meta_data
```

```python
population = population.merge(meta_data, how = 'inner', on = 'Country Name')
```
